```python
import numpy as np

import RBN_flip

K = 4
N= 5
N_agent = 5
N_environment = 7
r = 0.5
mutation_rate = 0.5
threshold = 0
d_r = 0.1
d_mutation = 0.1
num_T = 20
num_processes = 10
maxiter = 100
iteration_for_average = 20
#p = 0.5
iterations_convergence = 20
# Create an instance of the RBN class
network = RBN_flip.RBN(K,N,r)

```

Plotting the graph

```python
network.plot_states()
```

Printing the attractors and their probability

```python
RBN_flip.print_pmf(network, N)  
```

# Calculating Fisher Information

The cells below show proof that the Fisher information is calculated appropriately

Showing that the average of pmfs converges when the number used for the average is higher

```python
RBN_flip.convergence(num_T, N, network,r)
```

A bar plot, showing the probability of each state of the network for r = 0.1 and r = 0.5

Calculating the Fisher information for multiple r values
Also showing the value of the difference between pmfs for a d_r, as a check on whether the code behind the Fisher plot works. 

```python
RBN_flip.Fisher_plot(d_r, num_T, threshold, num_processes, network)
```

Showing how the Hellinger distance decreases over epochs for a certain mutation rate and different r

```python


change_array, zero_array, iteration_to_zero_array, rate_array, hellinger_distance_stack = RBN_flip.calculate_decrease_hellinger_distance_r(K, r, iterations_convergence, mutation_rate, N_agent, N_environment, d_mutation, maxiter, iteration_for_average, d_r, num_T, threshold,num_processes)

r_values = np.linspace(0, 1, len(change_array))
RBN_flip.plot_results_r(r_values, rate_array)
```

Showing how the Hellinger distance decreasesover epochs for a certain r and different mutation rates
